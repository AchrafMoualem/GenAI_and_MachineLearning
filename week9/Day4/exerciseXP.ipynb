# 🌟 Exercise 2: Ethical and Security Risks of Generative AI

**1. Deepfake Political Manipulation**

**Risks**

- People can be tricked by fake videos of politicians

- Voters may make decisions based on lies

- Trust in news and democracy decreases

**How to reduce the risk**

- Force platforms to label AI-generated videos

- Use tools that can detect deepfakes

**2. Synthetic Identity Fraud**

**Risks**

- Fake faces or voices can fool security systems

- Criminals can steal money or data

- Biometric systems lose credibility

**How to reduce the risk**

- Combine biometrics with extra security steps

- Use AI to detect fake identities

**3. Generative AI in Cyber Warfare**

**Risks**

- Governments may receive false intelligence

- Bad decisions can be made

- Conflicts may escalate

**How to reduce the risk**

- Keep humans in charge of critical decisions

- Create international rules for military AI use

**4. AI-Generated Malware**

**Risks**

- Malware becomes more advanced and harder to detect

- Hackers can attack faster

- Traditional antivirus tools fail

**How to reduce the risk**

- Use AI for cyber defense too

- Control access to powerful code-generating AI

**5. Copyright and Intellectual Property Theft**

**Risks**

- Creators lose control of their work

- AI copies existing styles too closely

- Legal problems increase

**How to reduce the risk**

- Require permission to use copyrighted data

- Add systems to track original content

# 🌟 Exercise 3: Optimization and Fine-Tuning of Generative AI Models

# Improving Generative AI Outputs

**1. Prompt Engineering**

Clear and detailed prompts lead to better results.

- **Image**
  - ❌ Generate an image of a futuristic city  
  - ✅ Generate a realistic night-time futuristic city with neon lights, flying cars, and green technology

- **Poem**
  - ❌ Write a poem about the future  
  - ✅ Write a short, hopeful poem about humanity’s future and technology

- **Music**
  - ❌ Create a song in the style of classical music  
  - ✅ Create a calm orchestral song inspired by classical music (Mozart-style)


**2. Bias and Fairness in Training Data**

**Problem:** Biased data produces biased outputs.

**Solutions:**
- Use diverse and balanced data sources  
- Remove extreme or one-sided content  
- Regularly audit model outputs


**3. Fine-Tuning for Medical Research**

**Steps:**
1. Collect peer-reviewed medical papers  
2. Clean and preprocess the data  
3. Fine-tune a pretrained model (transfer learning)  
4. Validate outputs with experts and factual checks


**4. Evaluating Generative AI Performance**

**Metrics:**
- BLEU / ROUGE – text similarity and summaries  
- Perplexity – text fluency  
- FID – image realism and quality


**5. Controlling Creativity and Coherence**

**Techniques:**
- Temperature: low = safe, high = creative  
- RLHF: improve quality using human feedback  
- Attention mechanisms: maintain logical consistency

# 🌟 Exercise 4: Evaluating the Trade-offs Between GANs and VAEs

#### 1. Synthetic Medical Image Generation
**Best choice: VAEs**

VAEs are safer and more stable for medical data.  
They learn a structured representation of images and reduce privacy risks, even if the images are less sharp than GANs.


#### 2. AI-Assisted Creative Writing
**Best choice: VAEs**

VAEs allow better control over style and creativity.  
They are easier to use for text generation than GANs and support smooth idea transitions.


#### 3. Anomaly Detection in Financial Transactions
**Best choice: VAEs**

VAEs learn normal behavior well.  
Unusual transactions are easy to detect using reconstruction error, making them reliable and explainable.


#### 4. High-Resolution Fashion Design
**Best choice: GANs**

GANs generate more realistic and detailed images.  
They are ideal when visual quality matters most, despite harder training.


#### 5. Data Augmentation for Autonomous Vehicles
**Best choice: GANs**

GANs create realistic driving scenes and rare conditions.  
VAEs can help with smooth variations, but GANs provide better realism.

# 🌟 Exercise 5: Advanced Latent Space Exploration in VAEs

#### 1. Visualizing Latent Space (MNIST)
To check if the latent space is good:
- Encode images and plot them in **2D or 3D**
- Same digits should form **clear clusters**

**Useful techniques:**
- *PCA* – quick overview
- *t-SNE / UMAP* – clearer class separation
- *K-means* – check how well digits cluster


#### 2. Interpolating Between Two Digits (3 → 8)
Steps:
- Encode both images into latent vectors  
- Interpolate between the two vectors (step by step)  
- Decode each step to get smooth image transitions

**Why VAEs work better than GANs**
- VAEs have a **continuous and structured latent space**
- GAN latent spaces are less smooth and harder to control


#### 3. Controlling Output Variability (KL Term)
- **Increase KL** → more diversity, less sharp images  
- **Decrease KL** → better quality, less variation  

**Real-world example:**
- **Medical imaging**: low variability for accuracy  
- **Art generation**: higher variability for creativity


#### 4. Disentangling Latent Features (Faces)
To change only one feature:
- Identify the latent dimension linked to that feature  
- Modify only that value (e.g., hair color)  
- Keep all other dimensions the same  

This allows **controlled and realistic edits**.


#### 5. PCA vs VAE (Quick Comparison)

- **Linearity**
  - PCA: linear
  - VAE: non-linear

- **Interpretability**
  - PCA: components easy to understand
  - VAE: more complex but flexible

- **Reconstruction**
  - PCA: limited
  - VAE: higher quality

- **Generative ability**
  - PCA: not generative
  - VAE: fully generative
