## 🌟 Exercise 1: Traditional vs. Modern NLP: Comparative Analysis
1) Complete the table:

| Aspect                   | Traditional NLP                                                   | Modern NLP                                                                    |
| ------------------------ | ----------------------------------------------------------------- | ----------------------------------------------------------------------------- |
| **Feature Engineering**  | Manual feature engineering (n-grams, POS tags, TF-IDF, rules)     | Automatic feature learning via neural networks                                |
| **Word Representations** | Sparse vectors (Bag-of-Words, TF-IDF)                             | Dense embeddings (Word2Vec, GloVe, FastText, contextual embeddings like BERT) |

| **Advantages**           | Interpretable, fast, low computational cost                       | High accuracy, captures context and semantics                                 |
| **Disadvantages**        | Limited context understanding, lower performance on complex tasks | Computationally expensive, requires large datasets and resources              |


2) Discuss: How did the shift to modern NLP impact scalability and efficiency?

- The move to modern NLP made systems much more scalable and efficient. Instead of hand-crafting features, models now learn meaningful representations directly from data, which saves time and improves performance. These dense representations also generalize better, so the same model can work across many tasks. Modern architectures like Transformers are designed to run in parallel, making them much faster when trained on GPUs or TPUs. Thanks to this hardware parallelism, very large models can be trained on huge datasets. Finally, transfer learning allows models to reuse knowledge from pretraining, reducing the need for large task-specific datasets and speeding up deployment.

### BERT
- Bidirectional encoder-only Transformer; trained with Masked Language Modeling (and historically Next Sentence Prediction).
- Application: Text classification, Named Entity Recognition (NER), question answering.
- Why: It understands context from both left and right, making it very strong for language understanding tasks.

### GPT
- TDecoder-only Transformer; trained as an autoregressive causal language model.
- Application: Text generation, chatbots, code generation, creative writing.
- Why: It generates fluent and coherent text by predicting the next token step by step.

### T5
- Encoder–decoder Transformer using a unified text-to-text framework.
- Application: Summarization, translation, question answering, text classification.
- Why: It treats every NLP task as a text generation problem, making it flexible and consistent across tasks.


## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- Benefits (explain each in your words):
  1) Improved generalization: Pretrained models learn broad language patterns from large corpora, so they perform well on new or unseen data, not just the examples they were trained on.
  2) Less labeled data: Because much of the language knowledge is learned during pretraining, only a small amount of labeled data remember is needed to adapt the model to a specific task.
  3) Faster fine-tuning: Instead of training a model from scratch, fine-tuning starts from a strong pretrained model, which converges faster and requires fewer computational resources.
  4) Transfer learning: Knowledge learned from one task or domain can be reused for others, allowing the same model to be applied to many NLP tasks with minimal changes.
  5) Robustness: Pretraining on diverse data helps models handle noise, variation in language, and different writing styles more reliably.

- Ethical concerns: Large NLP models can reflect and amplify biases present in their training data, generate misinformation or hallucinated content, be misused for spam or manipulation, and raise privacy concerns if sensitive data appears in training data.
- Mitigations: These risks can be reduced through careful data curation and filtering, applying differential privacy during training, using safety filters and content moderation, aligning models with human values through RLHF, and conducting regular audits and evaluations for bias, fairness, and misuse.


## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- Q/K/V flow :
Each word is projected into Query (Q), Key (K), and Value (V) vectors.
Queries are compared with Keys (dot product), a softmax turns scores into weights, and these weights are used to mix the Values into a context-aware representation.
- Why multiple heads help :
Each head learns different projections, so the model can capture different types of relations in parallel (syntax, meaning, long-range links).
- Example sentence :
  'The scientist thanked the assistant because she helped with the experiment.'
   One head may link "she" → "assistant" (coreference).
   Another head may link "helped" → "experiment" (action–object relation).

### Pre-training Objectives
- MLM vs CLM: MLM (Masked LM): predicts masked words using both left and right context → used in BERT.
CLM (Causal LM): predicts the next word using only past context → used in GPT.
- When to prefer MLM vs CLM: Use MLM for understanding tasks (classification, NER, QA).
Use CLM for generation tasks (chat, storytelling, code).
- NSP: Early BERT used Next Sentence Prediction to learn sentence relations, but many modern models drop it because it brings limited gains and simpler objectives work better.

### Transformer Model Selection
- Sentiment on reviews: Encoder-only — best for text understanding and classification.
- Conversational chatbot (creative responses): Decoder-only — designed for fluent, creative text generation.
- Technical document translation (EN→ES): Encoder–Decoder — encoder understands the source, decoder generates the target language.

### Positional Encoding
- Purpose: Adds word order information so the model knows which word comes first or later.
- Example failure without positions: “Dog bites man” and “Man bites dog” would look the same to the model.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- T = 0.2: Very low randomness → outputs are deterministic, repetitive, and conservative. the model strongly favors the highest-probability tokens.

- T = 1.5: High randomness → more diverse and creative outputs, but with higher risk of incoherence or factual errors.

- T = 1.0: Neutral setting → balanced trade-off between diversity and coherence.

### 2) Application Design
- Bedtime stories (creativity vs coherence): Use moderately high temperature (≈ 1.1–1.3) to encourage imagination and variation while keeping the story understandable and engaging.
- Financial report summaries (accuracy/reliability): Use low temperature (≈ 0.2–0.4) with greedy decoding or low top-k/top-p, to minimize hallucinations and ensure consistent, factual outputs.

### 3) Temperature & Bias
- Higher temperature increases randomness, which can surface latent biases by allowing lower-probability tokens to appear. Lower temperature suppresses this by sticking to the most likely, often more neutral phrasing—but it does not remove bias, it only hides it.
- *Example:* At high temperature, a model asked about “typical jobs” might generate gender-stereotyped roles more freely, while at low temperature it may default to safer, more generic descriptions.


### Quick Generation Demo Across Temperatures


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"
prompt = "In the future, artificial intelligence will change the way humans work by"

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print("--- Temperature:", T)
    print(out[0]["generated_text"])  # Inspect how style changes with T
```
