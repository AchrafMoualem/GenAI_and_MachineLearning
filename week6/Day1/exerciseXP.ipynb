1. age - the age of the applicant is necessary to know if we deal with an old enough applicant because older = more stable 
1. credit_score - This is the single most important predictor, as it's a proven numeric summary of a person's creditworthiness based on their entire financial history.

2. debt_to_income - Directly measures repayment capacity. High DTI (>43%) indicates someone may struggle to make payments even with good income.

3. annual_income - Higher verified income provides greater ability to repay, regardless of other factors.

4. num_delinquencies - Past late payments strongly predict future payment problems. Recent delinquencies are especially telling.

5. total_utilization - High credit card utilization (>30%) signals current financial stress and cash flow issues.

6. employment_length - Job stability correlates with income stability. Longer employment = lower default risk.

7. loan_amount - Larger loans represent greater financial burden and risk exposure.

8. term - longer term = more payement opportonities to default.

9. interest_rate - Higher rates often indicate higher perceived risk by the lender, and also increase the monthly payment burden.

#### Excloded with explain :
1. home_ownership - While relevant, home ownership status is already indirectly captured in the credit score and overall financial profile. Including it might unfairly disadvantage renters.

2. purpose - Loan purpose is often correlated with income level and credit score (e.g., debt consolidation loans go to borrowers with existing debt problems). Using it could create circular reasoning in the model.

3. term - The loan term is already implicitly considered in the monthly payment calculation, which is influenced by loan amount and interest rate. Longer terms mean lower monthly payments but higher total risk.

4. application_type - Individual vs. joint application status doesn't provide meaningful predictive power beyond what's already captured in income and debt ratios.

5. state & zip_code - Geographic data can lead to proxies for race and protected class discrimination (redlining). It's excluded for both fair lending compliance and because economic factors are better captured by national data.
