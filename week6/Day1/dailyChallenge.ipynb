#### Machine Learning Essentials

Machine Learning (ML) enables computers to learn patterns from data without explicit programming. For data analysts, ML automates complex pattern recognition, handles large datasets efficiently, and generates predictive insights that drive data-driven decision-making.

#### Industry Applications
- Healthcare: Predicting disease diagnosis from patient symptoms and medical history
- Finance: Detecting fraudulent transactions in real-time banking systems
- Retail: Recommending products based on customer browsing and purchase behavior

#### Types of Machine Learning
1. Supervised Learning
- Learns from labeled data to predict outcomes
- Email spam classification using labeled spam/not-spam examples

2. Unsupervised Learning
- Finds hidden patterns in unlabeled data
- Customer segmentation for targeted marketing campaigns

3. Reinforcement Learning
- Learns through trial-and-error with rewards/penalties
- Training autonomous vehicles to navigate traffic

#### ML Model Development Process
1. Undersand the problem : 
define what you want to solve.

2. Data collection : 
gether data and split it into training and testing sets.

3. Feature Selection : 
Identify and select relevant variables from raw data that best predict the target outcome, removing redundant or irrelevant information.

4. Model Selection :
Choose appropriate algorithm (e.g., linear regression, decision trees, neural networks) based on problem type, data characteristics, and performance requirements.

5. Model training :
feed training data into the model so it learns patterns

6. Model Evaluation :
Assess model performance using metrics (accuracy, precision, recall) on test data to ensure reliability and generalization to new data.

7. Loss functions :
mesure prediction error (MSE (Regression), Binary Crossentropy (Binary classification), Crossentropy (Multi-class classification))

8. Model optimisation :
adjust parameters if performance is unsatisfied
