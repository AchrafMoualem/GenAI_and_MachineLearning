```python
# What is the difference between New York and California in terms of sales and profit? (Compare the total sales and profit between New York and California.)
```

```python
# Compare New York and California
ny_data = df[df['State'] == 'New York']
ca_data = df[df['State'] == 'California']

ny_sales = ny_data['Sales'].sum()
ny_profit = ny_data['Profit'].sum()

ca_sales = ca_data['Sales'].sum()
ca_profit = ca_data['Profit'].sum()

print(f"New York - Sales: ${ny_sales:,.2f}, Profit: ${ny_profit:,.2f}")
print(f"California - Sales: ${ca_sales:,.2f}, Profit: ${ca_profit:,.2f}")
print(f"\nDifference in Sales: ${abs(ny_sales - ca_sales):,.2f}")
print(f"Difference in Profit: ${abs(ny_profit - ca_profit):,.2f}")
```

```python
from matplotlib import ticker as mticker
ny_customers = (
    df[df['State'] == 'New York']
      .groupby('Customer Name')
      .agg(Total_Sales=('Sales', 'sum'),
           Total_Profit=('Profit', 'sum'),
           Total_Orders=('Order ID', 'nunique'))
      .reset_index()
      .sort_values('Total_Sales', ascending=False)
)

top10_ny = ny_customers.head(10)

fig, ax = plt.subplots(figsize=(12, 6))
bar_colors = ['gold' if i == 0 else '#3498db' for i in range(len(top10_ny))]
bars = ax.barh(top10_ny['Customer Name'][::-1], top10_ny['Total_Sales'][::-1],
               color=bar_colors[::-1], edgecolor='white')

ax.set_title('Top 10 Customers in New York by Total Sales', fontsize=13, fontweight='bold')
ax.set_xlabel('Total Sales ($)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

for bar in bars:
    w = bar.get_width()
    ax.text(w + 200, bar.get_y() + bar.get_height()/2,
            f'${w:,.0f}', va='center', fontsize=9)

plt.tight_layout()
plt.show()

outstanding = ny_customers.iloc[0]
print(f"⭐ Outstanding NY Customer: {outstanding['Customer Name']}")
print(f"   Sales: ${outstanding['Total_Sales']:,.2f}  |  Profit: ${outstanding['Total_Profit']:,.2f}  |  Orders: {outstanding['Total_Orders']}")
print()
print(top10_ny.to_string(index=False))
```

```python
state_profit = (
    df.groupby('State')
      .agg(Total_Sales=('Sales', 'sum'),
           Total_Profit=('Profit', 'sum'))
      .reset_index()
)
state_profit['Profit_Margin'] = (state_profit['Total_Profit'] / state_profit['Total_Sales']) * 100
state_profit.sort_values('Total_Profit', ascending=False, inplace=True)


print('Top 5 Most Profitable States:')
print(state_profit.head(5).to_string(index=False))
print('\nBottom 5 Least Profitable States:')
print(state_profit.tail(5).to_string(index=False))
```

```python
city_metrics = (
    df.groupby('City')
      .agg(Total_Sales=('Sales', 'sum'),
           Total_Profit=('Profit', 'sum'))
      .reset_index()
)
city_metrics['Profit_Margin'] = (city_metrics['Total_Profit'] / city_metrics['Total_Sales']) * 100

top20_city_sales  = city_metrics.nlargest(20, 'Total_Sales')
top20_city_profit = city_metrics.nlargest(20, 'Total_Profit')

# ── Top 20 by Sales ───────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))
profit_colors = ['#27ae60' if p >= 0 else '#e74c3c' for p in top20_city_sales['Total_Profit']]
bars = ax.bar(top20_city_sales['City'], top20_city_sales['Total_Sales'],
              color='#3498db', edgecolor='white')
ax.set_title('Top 20 Cities by Total Sales', fontsize=13, fontweight='bold')
ax.set_ylabel('Total Sales ($)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.xticks(rotation=45, ha='right')
for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, h + 500,
            f'${h:,.0f}', ha='center', va='bottom', fontsize=7.5, rotation=45)
plt.tight_layout()
plt.show()

# ── Top 20 by Profit ──────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))
colors_p = ['#27ae60' if p >= 0 else '#e74c3c' for p in top20_city_profit['Total_Profit']]
bars = ax.bar(top20_city_profit['City'], top20_city_profit['Total_Profit'],
              color=colors_p, edgecolor='white')
ax.set_title('Top 20 Cities by Total Profit', fontsize=13, fontweight='bold')
ax.set_ylabel('Total Profit ($)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.xticks(rotation=45, ha='right')
for bar in bars:
    h = bar.get_height()
    offset = 200 if h >= 0 else -1500
    ax.text(bar.get_x() + bar.get_width()/2, h + offset,
            f'${h:,.0f}', ha='center', va='bottom', fontsize=7.5, rotation=45)
plt.tight_layout()
plt.show()

# ── Profitability scatter among top 20 sales cities ───────────────────────
fig, ax = plt.subplots(figsize=(12, 7))
sc = ax.scatter(top20_city_sales['Total_Sales'], top20_city_sales['Total_Profit'],
                c=top20_city_sales['Profit_Margin'], cmap='RdYlGn',
                s=120, edgecolors='black', linewidths=0.5)
plt.colorbar(sc, ax=ax, label='Profit Margin (%)')
for _, row in top20_city_sales.iterrows():
    ax.annotate(row['City'],
                xy=(row['Total_Sales'], row['Total_Profit']),
                xytext=(5, 5), textcoords='offset points', fontsize=8)
ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_title('Sales vs Profit — Top 20 Cities (colour = Profit Margin)', fontsize=13, fontweight='bold')
ax.set_xlabel('Total Sales ($)')
ax.set_ylabel('Total Profit ($)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.tight_layout()
plt.show()
```

```python
top20_customers = (
    df.groupby('Customer Name')
      .agg(Total_Sales=('Sales', 'sum'),
           Total_Profit=('Profit', 'sum'),
           Orders=('Order ID', 'nunique'))
      .reset_index()
      .nlargest(20, 'Total_Sales')
)

fig, ax = plt.subplots(figsize=(12, 8))
colors_cust = ['#27ae60' if p >= 0 else '#e74c3c' for p in top20_customers['Total_Profit']]
bars = ax.barh(top20_customers['Customer Name'][::-1],
               top20_customers['Total_Sales'][::-1],
               color=colors_cust[::-1], edgecolor='white')
ax.set_title('Top 20 Customers by Total Sales\n(Green = profitable, Red = loss-making)',
             fontsize=13, fontweight='bold')
ax.set_xlabel('Total Sales ($)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
for bar in bars:
    w = bar.get_width()
    ax.text(w + 200, bar.get_y() + bar.get_height()/2,
            f'${w:,.0f}', va='center', fontsize=9)
plt.tight_layout()
plt.show()

print(top20_customers.to_string(index=False))
```

```python
import numpy as np
customer_sales = (
    df.groupby('Customer Name')['Sales']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

total_sales_all = customer_sales['Sales'].sum()
customer_sales['Cumulative Sales'] = customer_sales['Sales'].cumsum()
customer_sales['Cumulative Sales %'] = customer_sales['Cumulative Sales'] / total_sales_all * 100
customer_sales['Customer Rank %'] = (np.arange(1, len(customer_sales)+1) / len(customer_sales)) * 100

threshold_20s = customer_sales[customer_sales['Customer Rank %'] <= 20].iloc[-1]

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(customer_sales['Customer Rank %'], customer_sales['Sales'],
       width=100/len(customer_sales), color='#3498db', alpha=0.5, label='Individual Sales')

ax2 = ax.twinx()
ax2.plot(customer_sales['Customer Rank %'], customer_sales['Cumulative Sales %'],
         color='#e74c3c', linewidth=2.5, label='Cumulative Sales %')
ax2.axhline(80, color='green', linestyle='--', linewidth=1.5, label='80% Sales Line')
ax2.axvline(20, color='orange', linestyle='--', linewidth=1.5, label='20% Customer Line')
ax2.scatter([threshold_20s['Customer Rank %']], [threshold_20s['Cumulative Sales %']],
            color='purple', zorder=5, s=80)
ax2.annotate(f"Top 20% customers\n→ {threshold_20s['Cumulative Sales %']:.1f}% of sales",
             xy=(threshold_20s['Customer Rank %'], threshold_20s['Cumulative Sales %']),
             xytext=(30, 55), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='purple'), color='purple')

ax.set_xlabel('Customers (Ranked by Sales, %)')
ax.set_ylabel('Sales ($)')
ax2.set_ylabel('Cumulative Sales (%)')
ax2.set_ylim(0, 105)
ax.set_title('Cumulative Sales Curve — Pareto Analysis (Customers & Sales)',
             fontsize=13, fontweight='bold')

lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines2, labels2, loc='lower right')
plt.tight_layout()
plt.show()

n20s = int(len(customer_sales) * 0.2)
top20pct_sales = customer_sales.head(n20s)['Sales'].sum()
print(f"Total customers            : {len(customer_sales)}")
print(f"Top 20% ({n20s} customers) : ${top20pct_sales:,.2f}  ({top20pct_sales/total_sales_all*100:.1f}% of sales)")
print(f"\n{'Pareto holds ✅' if top20pct_sales/total_sales_all >= 0.80 else 'Pareto does NOT perfectly hold — top 20% drive ' + f"{top20pct_sales/total_sales_all*100:.1f}%" + ' of sales'}")
```

```python
# ── Priority matrix: States ───────────────────────────────────────────────
state_matrix = state_profit.copy()
state_matrix['Sales_Rank']  = state_matrix['Total_Sales'].rank(ascending=False)
state_matrix['Profit_Rank'] = state_matrix['Total_Profit'].rank(ascending=False)

fig, ax = plt.subplots(figsize=(12, 8))
sc = ax.scatter(state_matrix['Total_Sales'], state_matrix['Total_Profit'],
                c=state_matrix['Profit_Margin'], cmap='RdYlGn',
                s=80, edgecolors='black', linewidths=0.4, alpha=0.85)
plt.colorbar(sc, ax=ax, label='Profit Margin (%)')

# Annotate notable states
highlight = ['California', 'New York', 'Texas', 'Washington', 'Ohio', 'Illinois', 'Pennsylvania']
for _, row in state_matrix[state_matrix['State'].isin(highlight)].iterrows():
    ax.annotate(row['State'],
                xy=(row['Total_Sales'], row['Total_Profit']),
                xytext=(6, 4), textcoords='offset points', fontsize=9, fontweight='bold')

ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
med_sales = state_matrix['Total_Sales'].median()
ax.axvline(med_sales, color='grey', linewidth=0.8, linestyle=':')

ax.set_title('State Priority Matrix: Sales vs Profit\n(colour = Profit Margin)',
             fontsize=13, fontweight='bold')
ax.set_xlabel('Total Sales ($)')
ax.set_ylabel('Total Profit ($)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Quadrant labels
xmax = state_matrix['Total_Sales'].max()
ymax = state_matrix['Total_Profit'].max()
ymin = state_matrix['Total_Profit'].min()
ax.text(xmax*0.6, ymax*0.85, 'High Sales\nHigh Profit\n✅ RETAIN', fontsize=9,
        color='green', ha='center', style='italic')
ax.text(xmax*0.05, ymax*0.85, 'Low Sales\nHigh Profit\n📈 GROW', fontsize=9,
        color='blue', ha='center', style='italic')
ax.text(xmax*0.6, ymin*0.7, 'High Sales\nLow Profit\n⚠️ OPTIMISE', fontsize=9,
        color='orange', ha='center', style='italic')
ax.text(xmax*0.05, ymin*0.7, 'Low Sales\nLow Profit\n🔴 DEPRIORITISE', fontsize=9,
        color='red', ha='center', style='italic')

plt.tight_layout()
plt.show()

# ── Summary table ─────────────────────────────────────────────────────────
print('=== MARKETING DECISION SUMMARY ===')
print('\n🏆 HIGH-PRIORITY STATES (High Sales + High Profit):')
hi_hi = state_matrix[(state_matrix['Total_Sales'] >= med_sales) & (state_matrix['Total_Profit'] > 0)]
print(hi_hi[['State', 'Total_Sales', 'Total_Profit', 'Profit_Margin']]
      .sort_values('Total_Profit', ascending=False).head(8).to_string(index=False))

print('\n⚠️  OPTIMISATION NEEDED (High Sales but Low/Negative Profit):')
hi_lo = state_matrix[(state_matrix['Total_Sales'] >= med_sales) & (state_matrix['Total_Profit'] <= 0)]
print(hi_lo[['State', 'Total_Sales', 'Total_Profit', 'Profit_Margin']]
      .sort_values('Total_Profit').head(5).to_string(index=False))
```

```python
# ── City priority matrix ──────────────────────────────────────────────────
city_all = (
    df.groupby('City')
      .agg(Total_Sales=('Sales', 'sum'),
           Total_Profit=('Profit', 'sum'))
      .reset_index()
)
city_all['Profit_Margin'] = (city_all['Total_Profit'] / city_all['Total_Sales']) * 100
top40_cities = city_all.nlargest(40, 'Total_Sales')

fig, ax = plt.subplots(figsize=(13, 8))
sc = ax.scatter(top40_cities['Total_Sales'], top40_cities['Total_Profit'],
                c=top40_cities['Profit_Margin'], cmap='RdYlGn',
                s=100, edgecolors='black', linewidths=0.4, alpha=0.85)
plt.colorbar(sc, ax=ax, label='Profit Margin (%)')

for _, row in top40_cities.iterrows():
    ax.annotate(row['City'],
                xy=(row['Total_Sales'], row['Total_Profit']),
                xytext=(5, 3), textcoords='offset points', fontsize=7.5)

ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_title('City Priority Matrix: Sales vs Profit (Top 40 Cities by Sales)',
             fontsize=13, fontweight='bold')
ax.set_xlabel('Total Sales ($)')
ax.set_ylabel('Total Profit ($)')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.tight_layout()
plt.show()

print('\n=== TOP 10 CITIES — PRIORITISE FOR MARKETING ===')
print(top40_cities[top40_cities['Total_Profit'] > 0]
      .sort_values('Total_Profit', ascending=False)
      .head(10)[['City', 'Total_Sales', 'Total_Profit', 'Profit_Margin']]
      .to_string(index=False))

print('\n=== CITIES NEEDING OPTIMISATION (Negative Profit) ===')
print(top40_cities[top40_cities['Total_Profit'] < 0]
      .sort_values('Total_Profit')
      .head(10)[['City', 'Total_Sales', 'Total_Profit', 'Profit_Margin']]
      .to_string(index=False))
```
